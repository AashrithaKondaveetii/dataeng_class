```python
import pandas as pd
import numpy as np
```

```python
file_path = '/content/sample_data/acs2017_census_tract_data.csv.gz'
data = pd.read_csv(file_path)
```

```python
file_path2 = '/content/sample_data/COVID_county_data.csv.gz'
data2 = pd.read_csv(file_path2)
```

```python
selected_columns = ['County', 'State', 'TotalPop', 'IncomePerCap', 'Poverty']
data_filtered = data[selected_columns].copy()

data_filtered['ID'] = range(len(data_filtered))

aggregated_data = data_filtered.groupby(['County', 'State']).agg({'TotalPop': 'sum', 'IncomePerCap': 'mean', 'Poverty': 'mean'})

County_Info = aggregated_data.reset_index()

County_Info.rename(columns={'County': 'county','TotalPop': 'Population', 'IncomePerCap': 'PerCapitaIncome'}, inplace=True)

County_Info['ID'] = County_Info.index + 1
County_Info
```

```python
counties = [
    ("Loudoun County", "Virginia"),
    ("Washington County", "Oregon"),
    ("Harlan County", "Kentucky"),
    ("Malheur County", "Oregon")
]

filtered_counties = County_Info[
    County_Info[['county', 'State']].apply(tuple, axis=1).isin(counties)
]

print(filtered_counties)
```

```python
most_populous_county = County_Info[County_Info['Population'] == County_Info['Population'].max()]

least_populous_county = County_Info[County_Info['Population'] == County_Info['Population'].min()]

print("Most populous county in the USA:\n", most_populous_county)
print("Least populous county in the USA:\n", least_populous_county)
```

```python
data2['ID'] = data2['county'] + '_' + data2['state']
data2['ID'] = pd.factorize(data2['ID'])[0] + 1
data2
```

```python
data2['date'] = pd.to_datetime(data2['date'], format='%Y-%m-%d')
data2['month'] = data2['date'].dt.to_period('M')
data2
```

```python
Covid_Monthly = data2.groupby(['ID', 'month']).agg({
    'cases': 'sum',
    'deaths': 'sum',
    'county': 'first'
}).reset_index()

Covid_Monthly['ID'] = range(1, len(Covid_Monthly) + 1)
Covid_Monthly = Covid_Monthly[['ID', 'month', 'cases', 'deaths', 'county']].copy()

Covid_Monthly.columns = ['ID', 'Month', 'cases', 'deaths', 'county']

Covid_Monthly
```

```python
print(Covid_Monthly[(Covid_Monthly['county'] == 'Malheur') & (Covid_Monthly['Month'] == '2020-08') ])
print(Covid_Monthly[(Covid_Monthly['county'] == 'Malheur') & (Covid_Monthly['Month'] == '2021-01') ])
print(Covid_Monthly[(Covid_Monthly['county'] == 'Malheur') & (Covid_Monthly['Month'] == '2021-02') ])
```

```python
covid_Summary = (Covid_Monthly.groupby("ID").agg({"cases": "sum", "deaths": "sum"}).reset_index())
covid_Summary.rename(columns={"cases": "TotalCases", "deaths": "TotalDeaths"}, inplace=True)
covid_Summary = pd.merge(covid_Summary, County_Info, how="left", on="ID")
covid_Summary["TotalCasesPer100K"] = round((covid_Summary["TotalCases"]) / (covid_Summary["Population"] / 100000), 2)
covid_Summary["TotalDeathsPer100K"] = round((covid_Summary["TotalDeaths"]) / (covid_Summary["Population"] / 100000), 2)
display(covid_Summary)
```

```python
counties = [('Washington County', 'Oregon'), ('Malheur County', 'Oregon'), ('Loudoun County', 'Virginia'), ('Harlan County', 'Kentucky')]

filtered_df = covid_Summary[
    (covid_Summary['county'].isin([county[0] for county in counties])) &
    (covid_Summary['State'].isin([county[1] for county in counties]))
]

poverty_percentage = filtered_df[['county', 'State', 'Poverty']]
total_cases_per_100k = filtered_df[['county', 'State', 'TotalCasesPer100K']]

print("Poverty Percentage:")
print(poverty_percentage)
print("\nTotal Cases per 100K:")
print(total_cases_per_100k)
```
