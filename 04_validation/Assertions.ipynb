## Existance Assertion

1. Every Crash ID has Crash date

```python
import pandas as pd

def print_crash_dates(csv_file_path):
    data = pd.read_csv(csv_file_path)
    date_columns = ['Crash Month', 'Crash Day', 'Crash Year']

    if data[date_columns].isnull().any().any():
        print("Warning: Some records are missing date information.")
    else:
        print("All records have complete date information.")

    for index, row in data.iterrows():
        crash_date = f"{int(row['Crash Month']):02d}/{int(row['Crash Day']):02d}/{int(row['Crash Year'])}" if not row[date_columns].isnull().all() else "Date not available"
        print(f"Crash ID: {row['Crash ID']}, Crash Date: {crash_date}")

csv_file_path = "/content/crash_data.csv"
print_crash_dates(csv_file_path)
```

2. Each crash of record type 1 has a Serial #

```python
import pandas as pd

def check_serial_numbers(csv_file_path):
    data = pd.read_csv(csv_file_path)

    type_1_records = data[data['Record Type'] == 1]

    missing_serials = type_1_records['Serial #'].isnull().any()
    if missing_serials:
        print("Warning: Some records of type 1 are missing a Serial #.")
    else:
        print("All records of type 1 have a Serial #.")

    for index, row in type_1_records.iterrows():
        serial = row['Serial #'] if pd.notnull(row['Serial #']) else "No Serial # available"
        print(f"Record Type: {row['Record Type']}, Crash ID: {row['Crash ID']}, Serial #: {serial}")

csv_file_path = "/content/crash_data.csv"
check_serial_numbers(csv_file_path)
```

2. All crash day numbers must be valid days of the month

```python
import csv

def is_valid_day(day):
    try:
        return 1 <= int(float(day)) <= 31
    except ValueError:
        return False

def retrieve_valid_crash_days(csv_file_path):
    valid_crash_days = []
    invalid_days = False
    try:
        with open(csv_file_path, mode='r', newline='', encoding='utf-8') as csvfile:
            csvreader = csv.DictReader(csvfile)
            for row in csvreader:
                crash_day = row['Crash Day'].strip()
                if is_valid_day(crash_day):
                    valid_crash_days.append(row)
                else:
                    invalid_days = True
                    print(f"Invalid day: {crash_day}")
    except Exception as e:
        print(f"Failed to read the file: {e}")
    return valid_crash_days, not invalid_days

csv_file_path = "/content/crash_data.csv"
valid_crashes, all_days_valid = retrieve_valid_crash_days(csv_file_path)

if all_days_valid:
    print("All crash day numbers in the data are valid days of the month.")
else:
    print("Some crash day numbers in the data are not valid days of the month.")


for crash in valid_crashes:
    crash_day = int(float(crash['Crash Day'].strip()))
    print(f"Crash ID: {crash['Crash ID'].strip()}, Crash Day: {crash_day}")
```

## Intra Record Assertion

1. If a crash record has a latitude coordinate recorded in seconds, it should also have a corresponding longitude coordinate recorded in seconds

```python
import csv

def is_valid_second(second):
    try:
        second = float(second)
        return 0 <= second < 60
    except ValueError:
        return False

def retrieve_valid_lat_long_seconds(csv_file_path):
    valid_records = []
    invalid_records = False
    try:
        with open(csv_file_path, mode='r', newline='', encoding='utf-8') as csvfile:
            csvreader = csv.DictReader(csvfile)
            for row in csvreader:
                lat_seconds = row['Latitude Seconds'].strip()
                long_seconds = row['Longitude Seconds'].strip()

                if is_valid_second(lat_seconds) and is_valid_second(long_seconds):
                    valid_records.append(row)
                else:
                    invalid_records = True
                    print(f"Invalid coordinates: Latitude Seconds: {lat_seconds}, Longitude Seconds: {long_seconds}")
    except Exception as e:
        print(f"Failed to read the file: {e}")

    return valid_records, not invalid_records

csv_file_path = '/content/sample_data/crash_data.csv'
valid_coordinates, all_seconds_valid = retrieve_valid_lat_long_seconds(csv_file_path)

if all_seconds_valid:
    print("All latitude and longitude seconds in the data are valid.")
else:
    print("Some latitude or longitude seconds in the data are not valid.")

for record in valid_coordinates:
    lat_seconds = record['Latitude Seconds'].strip()
    long_seconds = record['Longitude Seconds'].strip()
    print(f"Crash ID: {record['Crash ID'].strip()}, Latitude Seconds: {lat_seconds}, Longitude Seconds: {long_seconds}")
```

2. If a crash record has a latitude coordinate then it should also have a longitude coordinate

```python
import csv

def retrieve_valid_lat_long_degrees(csv_file_path):
    valid_records = []
    invalid_records = False
    try:
        with open(csv_file_path, mode='r', newline='', encoding='utf-8') as csvfile:
            csvreader = csv.DictReader(csvfile)
            for row in csvreader:
                lat_degrees = row['Latitude Degrees'].strip()
                long_degrees = row['Longitude Degrees'].strip()

                if lat_degrees and long_degrees:
                    valid_records.append(row)
                else:
                    invalid_records = True
                    print(f"Invalid or incomplete coordinate degrees: Latitude Degrees: {lat_degrees}, Longitude Degrees: {long_degrees}")
    except Exception as e:
        print(f"Failed to read the file: {e}")

    return valid_records, not invalid_records

csv_file_path = '/content/sample_data/crash_data.csv'
valid_coordinates, all_degrees_valid = retrieve_valid_lat_long_degrees(csv_file_path)

if all_degrees_valid:
    print("All records with a latitude also have a corresponding longitude")
else:
    print("Some records are missing a corresponding latitude or longitude")

for record in valid_coordinates:
    lat_degrees = record['Latitude Degrees'].strip()
    long_degrees = record['Longitude Degrees'].strip()
    print(f"Crash ID: {record['Crash ID'].strip()}, Latitude Degrees: {lat_degrees}, Longitude Degrees: {long_degrees}")
```

## Inter Record Assertion

1. Every vehicle listed in the crash data was part of a known crash

```python
import csv

def verify_vehicle_crash_data(vehicle_data_file_path, crash_data_file_path):

    with open(vehicle_data_file_path) as f:
        reader = csv.DictReader(f)
        vehicle_data = [(row['Crash ID'], int(float(row['Vehicle ID']))) for row in reader]

    with open(crash_data_file_path) as f:
        reader = csv.DictReader(f)
        crash_ids = {row['Crash ID'] for row in reader}

    for crash_id, vehicle_id in vehicle_data:
        if crash_id not in crash_ids:
            print(f"Vehicle {vehicle_id} was not part of a known crash.")
            return

    print("Every vehicle listed in the vehicle data was part of a known crash.")

    for crash_id, vehicle_id in vehicle_data:
        print(f"Crash ID: {crash_id}, Vehicle ID: {vehicle_id}")

vehicle_data_file_path = '/content/sample_data/vehicle_data.csv'
crash_data_file_path = '/content/sample_data/crash_data.csv'
verify_vehicle_crash_data(vehicle_data_file_path, crash_data_file_path)
```

3. Vehicles which are involved in a crash but not with an object at the time of the crash


```python
import csv

def find_vehicles_without_collision(crash_file_path, vehicle_file_path):

    crash_data = []
    with open(crash_file_path, 'r') as f:
        reader = csv.DictReader(f)
        for row in reader:
            crash_data.append(row)

    vehicle_data = []
    with open(vehicle_file_path, 'r') as f:
        reader = csv.DictReader(f)
        for row in reader:
            vehicle_data.append(row)

    vehicle_ids = set([int(float(row["Vehicle ID"])) for row in vehicle_data])

    for vehicle_id in vehicle_ids:
        has_collision = any([row["Crash Type"] == "8" and int(float(row["Vehicle ID"])) == vehicle_id for row in crash_data])
        if not has_collision:
            print(f"This Vehicle {vehicle_id} was involved in a crash but not with an object.")
        else:
          print(f"This Vehicle {vehicle_id} was involved in a crash but with an object")

crash_file_path = "/content/sample_data/crash_data.csv"
vehicle_file_path = "/content/sample_data/vehicle_data.csv"
find_vehicles_without_collision(crash_file_path, vehicle_file_path)
```

## Summary Assertion

1. CrashId is unique across all the records

```python
import csv

def check_unique_crash_id(csv_file_path):
    crash_ids = set()
    duplicate_ids = False
    try:
        with open(csv_file_path, mode='r', newline='', encoding='utf-8') as csvfile:
            csvreader = csv.DictReader(csvfile)
            for row in csvreader:
                crash_id = row['Crash ID'].strip()

                if crash_id in crash_ids:
                    duplicate_ids = True
                    print(f"Duplicate CrashId found: {crash_id}")
                else:
                    crash_ids.add(crash_id)
    except Exception as e:
        print(f"Failed to read the file: {e}")

    return not duplicate_ids

csv_file_path = '/content/sample_data/crash_data.csv'
is_unique = check_unique_crash_id(csv_file_path)

if is_unique:
    print("All CrashId present in the data are unique")
else:
    print("Some CrashId present in the data are not unique")
```

2. There are thousands of crashes not millions

```python
import csv

def count_unique_crash_id(csv_file_path):
    crash_ids = set()
    try:
        with open(csv_file_path, mode='r', newline='', encoding='utf-8') as csvfile:
            csvreader = csv.DictReader(csvfile)
            if 'Crash ID' not in csvreader.fieldnames:
                raise ValueError("No 'CrashId' column in the CSV file")
            for row in csvreader:
                crash_id = row['Crash ID'].strip()
                crash_ids.add(crash_id)
    except Exception as e:
        print(f"Failed to read the file: {e}")

    return len(crash_ids)

csv_file_path = '/content/sample_data/crash_data.csv'
unique_crashes_count = count_unique_crash_id(csv_file_path)

if unique_crashes_count < 1000000:
    print("There are thousands of crashes not millions")
else:
    print("There are millions of crashes not thousands")
```

3. The majority of crashes (50%) occurred from the afternoon to night commute hours of 2 PM to 11 PM

```python
import csv

def count_crashes_during_extended_hours(csv_file_path):
    total_crashes = 0
    extended_hour_crashes = 0
    try:
        with open(csv_file_path, mode='r', newline='', encoding='utf-8') as csvfile:
            csvreader = csv.DictReader(csvfile)
            for row in csvreader:
                total_crashes += 1
                crash_hour = int(float(row['Crash Hour'].strip()))
                if 14 <= crash_hour < 23:  # 2 PM to 11 PM
                    extended_hour_crashes += 1
    except Exception as e:
        print(f"Failed to read the file: {e}")
        return None, None

    return extended_hour_crashes, total_crashes

csv_file_path = '/content/sample_data/crash_data.csv'
extended_hour_crashes, total_crashes = count_crashes_during_extended_hours(csv_file_path)

if extended_hour_crashes is not None and total_crashes is not None:
    extended_hour_crashes_percentage = (extended_hour_crashes / total_crashes) * 100

    if extended_hour_crashes_percentage > 50:
        print("The majority of crashes occurred from afternoon 2PM to night 11PM")
    else:
        print("The majority of crashes did not occur from afternoon 2PM to night 11PM")
```

## Statistical Analysis

1. Crashes are more likely to occur on weekends than weekdays

```python
import csv
import matplotlib.pyplot as plt

def count_crashes_by_day(csv_file_path):
    weekday_crashes = 0
    weekend_crashes = 0
    try:
        with open(csv_file_path, mode='r', newline='', encoding='utf-8') as csvfile:
            csvreader = csv.DictReader(csvfile)
            for row in csvreader:
                weekday_code = int(float(row['Week Day Code'].strip()))
                if 2 <= weekday_code <= 6:  # Monday to Friday
                    weekday_crashes += 1
                elif weekday_code == 1 or weekday_code == 7:  # Sunday or Saturday
                    weekend_crashes += 1
    except Exception as e:
        print(f"Failed to read the file: {e}")
        return None, None

    return weekday_crashes, weekend_crashes

csv_file_path = '/content/sample_data/crash_data.csv'
weekday_crashes, weekend_crashes = count_crashes_by_day(csv_file_path)

if weekday_crashes is not None and weekend_crashes is not None:
    if weekend_crashes > weekday_crashes:
        print("Crashes are more likely to occur during weekends than weekdays")
    else:
        print("Crashes are more likely to occur during weekdays than weekends")

    # Plotting the data
    labels = ['Weekdays', 'Weekends']
    values = [weekday_crashes, weekend_crashes]
    plt.bar(labels, values, color=['lightgreen', 'skyblue'])
    plt.xlabel('Day Type')
    plt.ylabel('Number of Crashes')
    plt.title('Number of Crashes by Day Type')
    plt.show()
```

2. Crashes are evenly/uniformly distributed throughout the months of the year.

```python
import csv
import matplotlib.pyplot as plt
import numpy as np

def analyze_crashes(csv_file_path):
    month_names = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September',
    'October', 'November', 'December']

    crash_count = [0] * 12

    with open(csv_file_path) as f:
        reader = csv.DictReader(f)
        for row in reader:
            if row['Crash Month'] != '':
                month = int(float(row['Crash Month'])) - 1
                crash_count[month] += 1

    total_crashes = sum(crash_count)
    avg_crashes = total_crashes / 12


    print("Crashes by month:")
    for i, count in enumerate(crash_count):
        month_name = month_names[i]
        print(f"{month_name}: {count}")

    print(f"\nTotal number of crashes: {total_crashes}")
    print(f"Average number of crashes per month: {avg_crashes:.2f}")

    max_crashes = max(crash_count)
    min_crashes = min(crash_count)
    if max_crashes - min_crashes <= avg_crashes * 0.1:
        print("Crashes are evenly distributed throughout the months of the year.")
    else:
        print("Crashes are not evenly distributed throughout the months of the year.")

    plt.figure(figsize=(10, 6))
    bars = plt.bar(month_names, crash_count, color = 'skyblue')
    plt.xlabel('Month')
    plt.ylabel('Number of Crashes')
    plt.title('Number of Crashes by Month')
    plt.grid(axis='y')
    plt.xticks(rotation=45)
    plt.show()


csv_file_path = '/content/sample_data/crash_data.csv'
analyze_crashes(csv_file_path)
```

3. Crash frequency in winter months (December, January, February) is higher than in summer months (June, July, August).

```python
import csv
import matplotlib.pyplot as plt

def analyze_crashes(csv_file_path):
    crash_count = [0] * 12

    with open(csv_file_path) as f:
        reader = csv.DictReader(f)
        for row in reader:
            if row['Crash Month'] != '':
                month = int(float(row['Crash Month'])) - 1
                crash_count[month] += 1

    winter_months = [11, 0, 1]
    summer_months = [5, 6, 7]
    winter_crashes = sum(crash_count[i] for i in winter_months)
    summer_crashes = sum(crash_count[i] for i in summer_months)

    print(f"Number of crashes in winter months (December, January, February): {winter_crashes}")
    print(f"Number of crashes in summer months (June, July, August): {summer_crashes}")

    if winter_crashes > summer_crashes:
        print("Crash frequency in winter months is higher than in summer months")
    else:
        print("Crash frequency in summer months is higher than in winter months")

    plt.figure(figsize=(8, 5))
    seasons = ['Winter', 'Summer']
    crash_counts = [winter_crashes, summer_crashes]
    plt.bar(seasons, crash_counts, color=['skyblue', 'lightgreen'])
    plt.title('Frequency of Crashes in Summer vs Winter')
    plt.xlabel('Season')
    plt.ylabel('Number of Crashes')
    plt.grid(True, which='both', linestyle='--', linewidth=0.5)
    plt.show()

csv_file_path = '/content/sample_data/crash_data.csv'
analyze_crashes(csv_file_path)
```
